```python
import pandas as pd
import statistics
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('auto.csv')
df.head()
```

```python
df.columns
```

```python
print(statistics.mean(df['normalized-losses']))
print(statistics.median(df['normalized-losses']))
print(statistics.mode(df['normalized-losses']))
print(statistics.stdev(df['normalized-losses']))
```

```python
sns.heatmap(df.corr(),annot=True,cmap='RdYlGn',linewidths=0.2)
fig=plt.gcf()
fig.set_size_inches(12,12)
plt.title('Диаграмма корреляции')
plt.show()
```

```python
sns.displot(df['price'],kde=True)
plt.title('Price raspred')
plt.show()
```
